```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic.csv")
df.sample(5)
```

```python
df.info()
```

```python
# FOR CATEGORICAL DATA...
# 1. COUNT PLOT..
# df["Pclass"] = df["Pclass"].astype(str)
# df["Pclass"] is the integer type so it is necessary to convert it into string..
# df["Pclass"].value_counts().plot(kind = "bar")
sns.countplot(df['Sex'])
```

```python
# 2. Pie Chart..
df["Pclass"].value_counts().plot(kind = "pie", autopct = "%0.2f")
```

```python
# NUMERICAL DATA..
# 1. Histogram..
# Counting the data value in the interval count the number of people comes under the range of let say 0 - 10 is almost 66..
plt.hist(df["Age"], bins = 8)
plt.show()
```

```python
# 2. Dist Plot
# This gives the PDF function for a particular numeric col of a dataset tells the probability of the bins that people comes under or not..
# This also tells the skewness in the graph..
sns.distplot(df['Age'])
```

```python
# Box Plot..
# this gives the 5 value of the numeirc cols these are as follows..->
# 1. min = Q1 - 1.5*IQR
# 2. Q1 = 25th value of the col..
# 3. Q2 = median(50th value of the col..)
# 4. Q3 = 75th value of the data col..
# 5. max = Q3 + 1.5*IQR
# 6. IQR = Q3 - Q1 (TELLS THE data between 25 - 75)
# 7. above max and below min all the values are the outliers and the min and max value is located in the whiskers of the box plot of the data col..
plt.boxplot(df["Fare"])
plt.show()
```
