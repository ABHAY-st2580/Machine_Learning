```python
d['month'] = d['date'].dt.month
```

```python
d['month_name'] = d['date'].dt.month_name()
```

```python
d['day'] = d['date'].dt.day
```

```python
d['dayofweek'] = d['date'].dt.dayofweek
```

```python
d['day_name'] = d['date'].dt.day_name()
```
