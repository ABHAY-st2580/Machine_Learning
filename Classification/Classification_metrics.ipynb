```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples = 200, n_features = 2, n_informative = 1, n_redundant = 0, n_classes = 2, n_clusters_per_class = 1, hypercube = False, random_state = 42, class_sep = 1)
```

```python
plt.scatter(X[:,1], X[:,0], c = Y, s = 40)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
lor = LogisticRegression()
lor.fit(x_train, y_train)
```

```python
m = -(lor.coef_[0][0]/ lor.coef_[0][1])
b = -(lor.intercept_)/lor.coef_[0][1]
```

```python
m
```

```python
b
```

```python
x_new = np.linspace(-3, 3, 200)
y_new = m * x_new + b
```

```python
plt.plot(x_new, y_new, 'r-')
plt.scatter(X[:,0], X[:,1], c = Y)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
print(accuracy_score(lor.predict(x_test), y_test))
print(precision_score(lor.predict(x_test), y_test))
print(recall_score(lor.predict(x_test), y_test))
print(f1_score(lor.predict(x_test), y_test))
```

```python
cm = confusion_matrix(lor.predict(x_test), y_test)
```
