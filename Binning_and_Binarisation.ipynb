```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import KBinsDiscretizer, Binarizer
```

```python
df = pd.read_csv('Titanic.csv', usecols = ['Age', 'Fare', 'Survived'])
df.head(5)
df = df.dropna()
```

```python
# BINNING(DISCRETIZATION..)
def binning(bins, strategy, col):
    trf = KBinsDiscretizer(n_bins = bins, encode = 'ordinal', strategy = strategy)
    newdf = trf.fit_transform(df)
    newdf = pd.DataFrame(newdf, columns = df.columns)

    newdf.head(5)
    plt.figure(figsize = (14,5))
    plt.subplot(1,2,1)
    plt.hist(df[col])

    plt.subplot(1,2,2)
    plt.hist(newdf[col])

    plt.show()
```

```python
binning(10, 'quantile', 'Age')
```

```python
binning(10, 'uniform', 'Age')
```

```python
binning(10, 'quantile', 'Fare')
```

```python
binning(10, 'uniform', 'Fare')
```

```python
binning(10, 'kmeans', 'Age')
```

```python
binning(10, 'kmeans', 'Fare')
```

```python
# BINARIZATION..
df = pd.read_csv('Titanic.csv', usecols = ['Age', 'Fare'])
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
def binarizer(thresh, col):
    trf = Binarizer(threshold = thresh)
    newdf = trf.fit_transform(df)
    newdf = pd.DataFrame(newdf, columns = df.columns)

    newdf.head(5)
    plt.figure(figsize = (14,5))
    plt.subplot(1,2,1)
    plt.hist(df[col])

    plt.subplot(1,2,2)
    plt.hist(newdf[col])

    plt.show()
```

```python
binarizer(40, 'Age')
```

```python
binarizer(8, 'Fare')
```
